```python
import pandas as pd
import numpy as np
import urllib, json
import requests as _requests
import cgi
```

```python
header_data = {
    'Accept-Encoding': 'gzip, deflate, sdch',
    'Accept-Language': 'en-US,en;q=0.8',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64)'\
    ' AppleWebKit/537.36 (KHTML, like Gecko) Chrome/48.0.2564.82 '\
    'Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9'\
    ',image/webp,*/*;q=0.8',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive'
}
base_url = 'http://stats.nba.com/stats/'
```

```python
yr_string = '2016-17'
url_modifier = 'leaguegamelog'
api_params = {
    'LeagueID':'00',
    'Season':yr_string,
    'PlayerOrTeam':'T',
    'Direction':'DESC',
    'SeasonType':'Regular Season',
    'Sorter':'FGM'
}
pull_url = cgi.urlparse.urljoin(base_url, url_modifier)
response = _requests.get(pull_url, params=api_params,headers=header_data)
data = json.loads(response.content)
df = pd.DataFrame(data['resultSets'][0]['rowSet'])
df.columns = data['resultSets'][0]['headers']
schedule = df
game_ids = list(set(schedule['GAME_ID']))
game_ids = [int(x) for x in game_ids]
game_ids.sort()
game_ids = ['00'+str(x) for x in game_ids]
all_games = []
for game in game_ids:
    base_url = 'http://stats.nba.com/stats/'
    url_modifier = 'playbyplayv2'
    api_params = {
        'GameID':str(game),
        'StartPeriod':0,
        'EndPeriod':0
    }
    pull_url = cgi.urlparse.urljoin(base_url, url_modifier)
    response = _requests.get(pull_url, params=api_params,headers=header_data)
    data = json.loads(response.content)
    df = pd.DataFrame(data['resultSets'][0]['rowSet'])
    df.columns = data['resultSets'][0]['headers']
    pbp = df
    #Define helper function
    def fill_up(index,player_id,cols):
        index-=1
        open_count = sum(pbp.ix[index,cols]=='NA')
        while (pbp.ix[index,'EVENTMSGTYPE']!=12)&(sum(pbp.ix[index,cols]=='NA')>0 )& (sum(pbp.ix[index,cols]==player_id)==0):
            pbp.ix[index,cols[open_count-1]]=player_id
            index-=1
    home_cols = ['HOME_PLAYER_1','HOME_PLAYER_2','HOME_PLAYER_3','HOME_PLAYER_4','HOME_PLAYER_5']
    away_cols = ['AWAY_PLAYER_1','AWAY_PLAYER_2','AWAY_PLAYER_3','AWAY_PLAYER_4','AWAY_PLAYER_5']
    new_cols = home_cols+away_cols
    for col in new_cols:
        pbp[col]="NA"
    teams = [int(x) for x in list(set(pbp.ix[pbp.PERSON1TYPE==5,'PLAYER1_TEAM_ID']))]
    teams = teams + [int(x) for x in list(set(pbp.ix[pbp.PERSON1TYPE==4,'PLAYER1_TEAM_ID']))]
    for index,row in pbp.iterrows():
        #As long as not start of quarter
        if row.EVENTMSGTYPE!=12:
            pbp.ix[index,new_cols]=pbp.ix[index-1,new_cols]
        home_open = sum(pbp.ix[index,home_cols]=='NA')
        away_open = sum(pbp.ix[index,away_cols]=='NA')
        #EVENTMSGTYPE = 6 means foul
            #EVENTMSGACTIONTYPE =16 and 11 are events that I found to cause errors
        #EVENTMSGTYPE = 8 means substitution
        #EVENTMSGTYPE = 9 means timeout
        #EVENTMSGTYPE = 11 means 
        if (row.EVENTMSGTYPE!=8)&(row.EVENTMSGTYPE!=9)&(row.EVENTMSGTYPE!=11)&(not((row.EVENTMSGTYPE==6)&(row.EVENTMSGACTIONTYPE==16)))&(not((row.EVENTMSGTYPE==6)&(row.EVENTMSGACTIONTYPE==11)))& (not((row.EVENTMSGTYPE==9)&(row.EVENTMSGACTIONTYPE==2))):
            if not np.isnan(row.PLAYER1_TEAM_ID):
                if row.PLAYER1_TEAM_ID==teams[0]:
                    if sum(pbp.ix[index,away_cols]==row.PLAYER1_ID)<1:
                        pbp.ix[index,away_cols[away_open-1]]=row.PLAYER1_ID
                        away_open-=1
                        if away_open <0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,row.PLAYER1_ID,away_cols)
                else:
                    if sum(pbp.ix[index,home_cols]==row.PLAYER1_ID)<1:
                        pbp.ix[index,home_cols[home_open-1]]=row.PLAYER1_ID
                        home_open-=1
                        if home_open<0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,row.PLAYER1_ID,home_cols)
            if not np.isnan(row.PLAYER2_TEAM_ID):
                if row.PLAYER2_TEAM_ID==teams[0]:
                    if sum(pbp.ix[index,away_cols]==row.PLAYER2_ID)<1:
                        pbp.ix[index,away_cols[away_open-1]]=row.PLAYER2_ID
                        away_open-=1
                        if away_open <0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,row.PLAYER2_ID,away_cols)
                else:
                    if sum(pbp.ix[index,home_cols]==row.PLAYER2_ID)<1:
                        pbp.ix[index,home_cols[home_open-1]]=row.PLAYER2_ID
                        home_open-=1
                        if home_open<0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,row.PLAYER2_ID,home_cols)
            if not np.isnan(row.PLAYER3_TEAM_ID):
                if row.PLAYER3_TEAM_ID==teams[0]:
                    if sum(pbp.ix[index,away_cols]==row.PLAYER3_ID)<1:
                        pbp.ix[index,away_cols[away_open-1]]=row.PLAYER3_ID
                        away_open-=1
                        if away_open <0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,row.PLAYER3_ID,away_cols)
                else:
                    if sum(pbp.ix[index,home_cols]==row.PLAYER3_ID)<1:
                        pbp.ix[index,home_cols[home_open-1]]=row.PLAYER3_ID
                        home_open-=1
                        if home_open<0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,row.PLAYER3_ID,home_cols)
        elif (row.EVENTMSGTYPE==8):
            if not np.isnan(row.PLAYER1_TEAM_ID):
                player1 = row.PLAYER1_ID
                player2 = row.PLAYER2_ID
                if row.PLAYER1_TEAM_ID==teams[0]:
                    if sum(pbp.ix[index,away_cols]==player1)>0:
                        pbp.ix[index,pbp.ix[index,away_cols][pbp.ix[index,away_cols]==player1].index[0]]=player2
                    else:
                        pbp.ix[index,away_cols[away_open-1]]=player2
                        away_open-=1
                        if away_open <0:
                            print 'Error at game number ' + str(game) +' on play ' + str(index)
                        fill_up(index,player1,away_cols)
                else:
                    if sum(pbp.ix[index,home_cols]==player1)>0:
                        pbp.ix[index,pbp.ix[index,home_cols][pbp.ix[index,home_cols]==player1].index[0]]=player2
                    else:
                        pbp.ix[index,home_cols[home_open-1]]=player2
                        home_open-=1
                        if home_open <0:
                            print 'Error at game number ' + str(game)+ ' on play ' + str(index)
                        fill_up(index,player1,home_cols)
    col_counts = []
    for col in new_cols:
        col_counts.append(sum(pbp[col]=='NA'))
    for index,col in enumerate(col_counts):
        if col>sum(pbp.EVENTMSGTYPE==12):
            for period in list(set(pbp.ix[(pbp[new_cols[index]]=='NA')&(pbp.EVENTMSGTYPE!=12),'PERIOD'])):
                team = teams[0]
                team_cols = away_cols
                if index<5:
                    team = teams[1]
                    team_cols = home_cols
                start_index = min(pbp.ix[(pbp.PERIOD==period),:].index)+1
                url_modifier = 'boxscoretraditionalv2'
                api_params = {
                    'GameID':str(game),
                    'StartPeriod':period,
                    'EndPeriod':period,
                    'StartRange':0,
                    'EndRange':0,
                    'RangeType':1
                }
                pull_url = cgi.urlparse.urljoin(base_url, url_modifier)
                response = _requests.get(pull_url, params=api_params,headers=header_data)
                data = json.loads(response.content)
                df = pd.DataFrame(data['resultSets'][0]['rowSet'])
                df.columns = data['resultSets'][0]['headers']
                mins = '5:00'
                if period<5:
                    mins = '12:00'
                played_all = df.ix[(df.MIN==mins)&(df.TEAM_ID==team)]
                missing_player = 'NA'
                for player_id in played_all.PLAYER_ID: 
                    if player_id not in list(pbp.ix[start_index,team_cols]):
                           missing_player = player_id
                if missing_player != 'NA':
                    pbp.ix[(pbp.PERIOD==period)&(pbp.EVENTMSGTYPE!=12),new_cols[index]]=missing_player
                else:
                    print('Error - need to fix minutes: '+str(game))
    col_counts = []
    for col in new_cols:
        col_counts.append(sum(pbp[col]=='NA'))
    if sum(col_counts)!=sum(pbp.EVENTMSGTYPE==12)*10:
        print('Error - need to fix minutes: '+str(game))
    all_games.append(pbp)
```

```python
all_concat = pd.concat(all_games)
```

```python
all_concat.to_csv('pbp_2016.csv',index=False)
```
